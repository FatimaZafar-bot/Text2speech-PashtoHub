```python
import pandas as pd
import json
import os
import os
import json
import shutil
import librosa
import soundfile as sf
from google.colab import drive
drive.mount('/content/drive')
```

```python
from pathlib import Path
import re
```

```python
DATA_PATH='/content/drive/MyDrive/audios'
SENTENCES_PATH=f'/content/sentences.json'
AUDIO_FOLDER=f'{DATA_PATH}'
OUTPUT_PATH=f'{DATA_PATH}/pashto_dataset'
```

```python
with open(SENTENCES_PATH, 'r', encoding='utf-8') as file:
    data = json.load(file)

sentences_data = data[2]['data']
df = pd.DataFrame(sentences_data)

os.makedirs(f'{OUTPUT_PATH}', exist_ok=True)
os.makedirs(f'{OUTPUT_PATH}/wavs', exist_ok=True)


audio_files = [f for f in os.listdir(AUDIO_FOLDER) if f.endswith('.wav')]
print(f"Found {len(audio_files)} audio files in {AUDIO_FOLDER}")


matched_files = {}

for audio_file in audio_files:
    number_match = re.search(r'(\d+)', audio_file)
    if number_match:
        file_number = number_match.group(1)
        matching_rows = df[df['sentence_id'] == file_number]
        if not matching_rows.empty:
            matched_files[file_number] = audio_file
            print(f"Matched {audio_file} to sentence ID {file_number}")


print(f"Successfully matched {len(matched_files)} audio files with sentences")
print(f"Total sentences in JSON: {len(df)}")


unmatched_sentences=df[~df['sentence_id'].isin(matched_files.keys())]
if not unmatched_sentences.empty:
    print(f"Warning: {len(unmatched_sentences)} sentences have no matching audio file")
    print("First few unmatched sentence IDs:", unmatched_sentences['sentence_id'].head().tolist())

with open(f'{OUTPUT_PATH}/metadata.csv', 'w', encoding='utf-8') as metadata_file:
    for sentence_id, audio_file in matched_files.items():
        sentence_row = df[df['sentence_id'] == sentence_id].iloc[0]
        sentence = sentence_row['sentence']

        source_path = os.path.join(AUDIO_FOLDER, audio_file)
        target_path = f'{OUTPUT_PATH}/wavs/{sentence_id}.wav'
        shutil.copy(source_path, target_path)
        metadata_file.write(f"{sentence_id}|{sentence}\n")

print("Dataset preparation complete!")
```

```python
audio_files=sorted([f for f in os.listdir(AUDIO_FOLDER) if f.endswith('.wav')])
sentence_ids=sorted(df['sentence_id'].tolist())

print(f"Audio files count: {len(audio_files)}")
print(f"Sentence count: {len(sentence_ids)}")

with open(f'{DATA_PATH}/audio_sentence_mapping.txt', 'w', encoding='utf-8') as f:
    f.write("# Format: audio_filename|sentence_id\n")
    f.write("# Edit this file to correct any mismatches\n\n")
    for i in range(min(len(audio_files), len(sentence_ids))):
        f.write(f"{audio_files[i]}|{sentence_ids[i]}\n")

print(f"Created mapping template at {DATA_PATH}/audio_sentence_mapping.txt")
print("Please edit this file to correct any mismatches, then run the next processing step.")
```

```python
mapping = {}
with open(f'{DATA_PATH}/audio_sentence_mapping.txt', 'r', encoding='utf-8') as f:
    for line in f:
        if line.startswith('#') or not line.strip():
            continue
        audio_file, sentence_id = line.strip().split('|')
        mapping[audio_file] = sentence_id

os.makedirs(f'{OUTPUT_PATH}/wavs', exist_ok=True)

with open(f'{OUTPUT_PATH}/metadata.csv', 'w', encoding='utf-8') as metadata_file:
    for audio_file, sentence_id in mapping.items():
        matching_rows = df[df['sentence_id'] == sentence_id]
        if matching_rows.empty:
            print(f"Warning: No sentence found for ID {sentence_id}")
            continue

        sentence = matching_rows.iloc[0]['sentence']
        source_path=os.path.join(AUDIO_FOLDER, audio_file)
        target_path=f'{OUTPUT_PATH}/wavs/{sentence_id}.wav'
        shutil.copy(source_path, target_path)

        metadata_file.write(f"{sentence_id}|{sentence}\n")

print("Dataset preparation complete using manual mapping!")
```
